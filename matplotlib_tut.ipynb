```python
fig,axis=plt.subplots(figsize=(20,20),nrows=2,ncols=1)
# first
axis[0].scatter(batters["avg"],batters["strike_rate"],color="red",label="runs and average ratio")
axis[0].set_title("average and strike rate")
axis[0].set_xlabel("average ")

axis[0].set_ylabel("strike rate ")

# second
axis[1].scatter(batters["avg"],batters["runs"])
axis[1].set_title("average and runs")

axis[1].set_xlabel("average ")

axis[1].set_ylabel("runs")
```

## Surface Plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
xx.shape
yy.shape
z=xx**2+yy**2
z.shape
```

```python
xx
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.subplot(projection="3d")
res=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(res)
```
