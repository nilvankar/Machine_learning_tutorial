## most runs against 3 team by virat kohli

```python
vk=deliveries[deliveries["batsman"]=="V Kohli"]
```

```python

vk.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).head()
```

```python
def mostRunsTeam(name_of_batsman):
    bt=deliveries[deliveries["batsman"]==name_of_batsman]
    return bt.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).head()
```

```python
mostRunsTeam("MS Dhoni")
```

## isin() function

```python
death_overs=deliveries[deliveries["over"]>15]
```

```python
mostball=death_overs.groupby("batsman")["batsman_runs"].count()
```

```python
x=mostball>200
```

```python
destructive_batsman=mostball[x].index.tolist()
```

```python
final=deliveries[deliveries["batsman"].isin(destructive_batsman)]
```

```python
runs=final.groupby("batsman")["batsman_runs"].sum()
```

```python
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
strike_rate=(runs/balls)*100
```

```python
strike_rate.head()
```
