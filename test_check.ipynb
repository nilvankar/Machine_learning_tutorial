```python
batsman_records=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
batsman_records=batsman_records.head(100)
```

```python
x=np.linspace(-20,50,10000)
y=np.linspace(-20,50,10000)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.subplot(projection="3d")
res=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(res)
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.subplot()
res=ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(res)
```
