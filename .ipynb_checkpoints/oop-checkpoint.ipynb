```python
class Parent:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def person(self):
        print(f"Your name is {self.name} and you are {self.age} years old")
class Child(Parent):
    def __init__(self,name,age):
        self.name=name
        self.age=age
    # def person(self):
    #     print(f"Your name is {self.name} and you are {self.age} years old from child")
p1=Parent("Neel",12)
p1.person()
c1=Child("Abhinav",13)
c1.person()
```
