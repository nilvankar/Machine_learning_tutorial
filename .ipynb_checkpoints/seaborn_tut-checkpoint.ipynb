# Matrix Plot
* heatmap
* cluster map

```python
countries=pd.read_csv('life_exp.csv')

developing=countries[countries["Status"]=="Developing"]
developed=countries[countries["Status"]=="Developed"]
countries.columns
```

```python
grid=countries.pivot(index="Country",columns="Year",values="Life expectancy")
```

```python
sns.heatmap(grid)
plt.figure(figsize=(20,20))
# only figure level for heatmap
```

```python
devgrid=developing.pivot(index="Country",columns="Year",values="Life expectancy")
devolepedgrid=developed.pivot(index="Country",columns="Year",values="Life expectancy")
```

```python
sns.heatmap(devgrid,annot=True)
```

```python
sns.heatmap(devolepedgrid,annot=True,cmap="spring")
plt.figure(figsize=(20,20))
```

## categorical estimate
* bar plot
* point plot
* count plot
* find cental tendency
  

```python
sns.stripplot(data=tips,x="day",y="total_bill",jitter=False)
# axis level function
# most bill pay on sunday
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="strip")

# most bill pay on sunday
```

```python
# jitter
sns.catplot(data=tips,x="day",y="total_bill",jitter=0.75,kind="strip")

# most bill pay on sunday
```

## Swarm plot
* distribution

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm",palette='dark:red',hue="sex")

# most bill pay on sunday
```

## box plot
* univariant and bivariant analysis both
* five number summary mean,minimum,maximum,upper quartile75% lower quartile25%

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="box",hue="sex")
```

```python
sns.boxplot(data=tips,y="tip")
```

## voilin plot (box plot+kde plot)

```python
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex",split=2)
```

```python
sns.barplot(data=tips,x="day",y="total_bill",hue="sex")

# black line represents error bar if something repeating for example male and female
#bootstraping
```

```python
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",errorbar=None,kind="bar")
```

```python
sns.pointplot(data=tips,x="sex",y="total_bill",estimator="mean",hue="smoker",)

# black line represents error bar if something repeating for example male and female
#bootstraping
```
