```python
import matplotlib.pyplot as plt
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
matches=pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries.csv')
```

```python
ipl=deliveries.merge(matches,left_on="match_id",right_on="id")
```

```python
ipl.columns
```

```python

# Assuming 'ipl' is your DataFrame containing IPL data

# Step 1: Identify the top 50 batters by total runs scored
top50_batters = ipl.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False).head(50).index.tolist()

# Step 2: Filter the DataFrame to include only the top 50 batters
top50_data = ipl[ipl['batter'].isin(top50_batters)]

# Step 3: Calculate total runs for each batter
runs = top50_data.groupby("batter")["batsman_runs"].sum()

# Step 4: Calculate total balls faced for each batter
# Assuming 'ball_faced' is a column in the DataFrame. If not, you might need to calculate it based on the data available.
balls_faced = top50_data.groupby("batter")["batsman_runs"].count()

# Step 5: Calculate strike rate for each batter
strike_rate = (runs / balls_faced) * 100

# Display the strike rates
print(strike_rate)
sr=strike_rate.reset_index()
```
